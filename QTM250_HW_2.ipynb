# **The Babies of Bohemian Rhapsody: Did the Name "Rami" Become More Popular After 2018?**

#Introduction
Looking at data of names of people born each year can tell us a lot about the state of the country and the world during certain periods of time. In a broad sense, the number of children born each year can tell us about the surrounding circumstances of that year. Perhaps some years have less children born because of economic recessions or wars occurring during those years. Narrowing in on name data to look at specific names can tell us about trends in baby names over time and possibly even about the cultural influences that could have elicited those trends.

Media is a large part of culture and can affect many aspects of people's lives, even including what they choose to name their children. Movies in particular can serve as pinpoints of specific years, so we were interested in investigating whether popular movies actually influence baby names during and after the year they are released. 

In order to examine a clear relationship between a movie's release and the popularity of a certain baby name associated with that movie, we wanted to choose an actor's name that is relatively uncommon. That way, if we did find an increase in the name's usage, we could be more certain that the increase could be attributed to the movie as opposed to the name just being popular in general. With that in mind, we chose the popular movie *Bohemian Rhapsody* released in 2018 since that was the movie that brought actor Rami Malek to fame. Since "Rami" is a relatively uncommon name, we thought it would be a good fit for this assignment.


#Hypothesis:

Due to the popularity of *Bohemian Rhapsody*, we expect to see an increase in the use of the baby name "Rami" during and after 2018.
