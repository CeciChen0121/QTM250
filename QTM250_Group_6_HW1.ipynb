# **Emory Entertains: How Many Hours a Week Do Emory Students Spend on Entertainment?** 

# Introduction
As Emory students, we are often swamped with classes, extracurriculars, and homework, which does not often leave much time for fun, especially during the week. Once students declare their majors and start contemplating their future careers, it seems that they are on a race to the finish line of graduation, accomplishing as much academically and professionally as they can. However, technology is so pervasive throughout our lives that no matter how much time we devote to our studies, it seems that our time spent on entertainment (social media and movies/TV shows apps) somehow matches or exceeds our studying time.

We were interested in seeing whether the year that Emory students are in impacts the number of hours per week that they spend on entertainment. Most Emory students are busy, but are there certain years that devote more time to entertainment? Why might that be? Ultimately, the objective of this experiment was to determine which year spends the most time on entertainment to see if the results align with what we would expect based on our knowledge of the events and milestones that occur during each year as an Emory student.

Our thought process was that different years entail different amounts of work and thus certain years might provide more time for entertainment than others. More generally, we also wanted to get a sense of the overall average number of hours spent on entertainment per week across Emory students. 

A [2019 report](https://www.commonsensemedia.org/sites/default/files/uploads/research/2019-census-8-to-18-full-report-updated.pdf) by Common Sense Media about media use by teens and tweens found that the average amount of entertainment screen media teens consume per day is about 7 hours and 22 minutes, or about 52 hours per week. While not all college students are teenagers, this study was conducted when most current Emory students were teenagers, so we thought it would be interesting to compare the average number of hours of entertainment collected when Emory students were teenagers to the average now that they are college students. Also, the study is fairly recent, so we would expect to see a similar average in our own study, even though the sample size is much smaller.

The null hypothesis of our study is that all years at Emory spend the same amount of time on entertainment per week. However, we expect that there will be differences in the amount of time different years devote to entertainment based on the associated workload and free time of each year.

# Hypothesis:

The null hypothesis of our study is that all years at Emory spend the same amount of time on entertainment per week. However, we expect that there will be differences in the amount of time different years devote to entertainment based on the associated workload and free time of each year. Our Alternative hypothesis is that there will be a difference between the different years as we expect that the years with the highest average number of hours spent on entertainment per week are first-years and seniors, meaning sophomores and juniors will have the lowest average number of hours spent on entertainment per week. We expect this because first-yers are still new to college so they are not taking many classes. Additionally, Emory recommends their first-year students to take a lighter workload in their first year (between 15-17 credit hours), and will not allow a first-year student to take more than 19 credits in their first semester of college in order to adjust to college life. Thus, because of their lighter workload we assume they will spend more time on entertainment. Seniors on the other hand commonly have completed majority of their Major core classes by their fourth-year and will choose to take a lighter workload to enjoy their last year in college. While sophomores and juniors are in the years were they are discovering what field to major in as well as taking the required core classes in order to have a lighter workload in their senior year.
